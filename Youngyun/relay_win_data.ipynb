```python
import pandas as pd

df = pd.read_csv('2024_일자별순위.csv')
df.head()
```

```python
hanwha_df = df[df['팀명'] == '한화']
hanwha_df
```

```python
filtered_df = hanwha_df[df['연속'].str.endswith('승')]
filtered_df
```

```python
import pandas as pd

# 예제 데이터프레임
# hanwha_df = pd.DataFrame({
#     '연속': ['1승', '2승', '3승', '승', '4승', '승리'],
#     '기타열': [1, 2, 3, 4, 5, 6]
# })

# '연속' 열에서 '승'으로 끝나는 경우 필터링
filtered_df = hanwha_df[hanwha_df['연속'].str.endswith('승')]

# '연속' 열에서 '승' 앞의 숫자를 추출
filtered_df['승수'] = filtered_df['연속'].str.extract(r'(\d+)승', expand=False)

# '승수' 열을 정수형으로 변환, NaN 값을 0으로 대체
filtered_df['승수'] = filtered_df['승수'].fillna('0').astype(int)

# 승수가 2 이상인 경우만 필터링
result_df = filtered_df[filtered_df['승수'] >= 2]

print(result_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 예제 데이터프레임 (날짜 데이터와 승수 데이터 포함)
# filtered_df = pd.DataFrame({
#     '연속': ['2승', '3승', '4승'],
#     '승수': [2, 3, 4],
#     '날짜': pd.date_range(start='2024-01-01', periods=3, freq='D')
# })

# 날짜 열을 datetime 형식으로 변환 (필요시)
filtered_df['일자'] = pd.to_datetime(filtered_df['일자'])

# 꺾은선 그래프를 그리기 위한 데이터 준비
x = filtered_df['일자']
y = filtered_df['승수']

# 꺾은선 그래프 시각화
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='tab:blue')
plt.title('날짜에 따른 승수 변화')
plt.xlabel('일자')
plt.ylabel('승수')
plt.grid(True)
plt.xticks()  # 날짜 레이블이 겹치지 않도록 회전
plt.tight_layout()  # 레이아웃 조정
plt.show()
```
